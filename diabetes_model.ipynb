# 2. Analyse the effect of age on Outcome using Boxplot

```python
sns.boxplot(x='Outcome',y='Age',data=data)
```

# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.stripplot(y=data['DiabetesPedigreeFunction'],x=data['Outcome'],data=data)
plt.grid()
```

```python

import matplotlib.pyplot as plt

sns.scatterplot(x=data['Insulin'],y=data['Glucose'],data=data,hue=data['Outcome'])
```
